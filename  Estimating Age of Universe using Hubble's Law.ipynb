```python
import pandas as pd
import numpy as np
df=pd.read_csv('https://raw.githubusercontent.com/Arhaan/CABKrittikaTutorials/main/Selection_Assignment/data.txt')
array1 = df['mod0'].to_numpy()
array2= df['vgsr'].to_numpy()
B=(array1/5)-5
xdata=np.power(10,B)
print(xdata)
```

```python
%matplotlib inline
# import all the important packages

import numpy as np
from scipy.optimize import curve_fit as cf
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

```python
# Define a new function
def linear_func(x,a,b):
    return a*x + b
#ydata = linear_func(xdata,70,52) + np.random.randn(len(xdata))
ydata=array2
plt.scatter(xdata,ydata,s=1)
p_opt, p_cov = cf(linear_func,xdata,ydata)

plt.plot(xdata,linear_func(xdata,*p_opt),label='Best Fit',color='red')
plt.scatter(xdata,ydata,label='given Data',s=1,color='pink')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Fit')
plt.legend()
# compare the original parameters with the estimated ones
#print('a_slope={}, b_intercept={}'.format(*(abs(p_opt-[70,52]))))

# The 'format' function replaces the curly brackets with the arguments provided to it. For example 
# print('a={}, b={}'.format(1,2)) will print 'a=1,b=2'.



slope, intercept, r_value, p_value, std_err = linregress(xdata,array2)
print(f"Hubbles constant={slope}km/s/mpc")
print(f"The age of the universe={1/slope}mpc/km/s")
plt.scatter(x = xdata, y = array2)
plt.xlabel("distance to galaxy")
plt.ylabel("line of slight velocity")
plt.show()
```
