```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('https://raw.githubusercontent.com/Arhaan/CABKrittikaTutorials/main/Selection_Assignment/stardata.csv')

df.loc[-1]=[56.56677415601188 ,24.189734528215197 ,7.3935352662039655 ,8.208464]
df.index=df.index+1
df=df.sort_index()
headerList = ['RA', 'declination', 'parallax', 'apparent magnitude']
df.to_csv("gfg2.csv", header=headerList, index=False)
df = pd.read_csv("gfg2.csv")
df.drop(df[df['parallax']=='--'].index, inplace= True)

df[['parallax']]=df[['parallax']].apply(pd.to_numeric)
# d=1/parallax
dis=1/df['parallax']
df.insert(4,'Distance',dis)
#Absolute magnitude=m+5-5log(d)
M=df['apparent magnitude']+5-5*np.log10(df['Distance'])
df.insert(5,'Absolute magnitude',M)
df.hist(column='Absolute magnitude',bins=50,figsize=(13,16))


print(df)
    

```

```python
size = df['apparent magnitude'].to_numpy()
s = [(12-s)*10 for s in size]


fig,ax = plt.subplots(figsize=(17, 11))
ax.scatter(x = df['RA'], y = df['declination'],s=s)
plt.xlabel("RA")
plt.ylabel("declination")

plt.show()
```
